```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anm
from IPython import display
```

```python
fig = plt.figure()
ims = []

for i in range(10):
    rand = np.random.randn(100)
    im = plt.plot(rand)
    ims.append(im)
    
ani = anm.ArtistAnimation(fig, ims, interval=100)
html = display.HTML(ani.to_jshtml())
display.display(html)
plt.close()
```

```python
fig, ax = plt.subplots()

line, = ax.plot([])

ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.1, 1.1)

def animate(frame_num):
    x = np.linspace(0,6,1000)
    y = np.sin(x + 2*np.pi * frame_num/100)
    line.set_data((x, y))
    return line

anim = anm.FuncAnimation(fig, animate, frames=100, interval=50)
html = display.HTML(anim.to_jshtml())
display.display(html)
plt.close()
```
